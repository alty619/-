```python
# 데이터 프레임의 꽃
# apply 함수
# 활용
# 본인이 원하는 행과 열에 연산 혹은 function을 적용할 수 있다.
#열 기준 집계 : axis=0
#행 기준 집계 : axis=1
```

```python
# 열 기준 집계_ axis=0은 row, axis=1은 column을 나타낸다고 생각하면, diameter의 전체 행들의 average를 집계한 것이다.
abalone_df[['diameter']].apply(np.average, axis=0)
```

```python
# 행 기준 집계
abalone_df[['diameter']].apply(np.average,axis=1)
```

```python
# 사용자 함수를 통한 집계 ********** 중요
import math
def avg_ceil(x,y,z):
  return math.ceil((x+y+z)) # ceil은 올림.

abalone_df[['diameter','height','whole_weight']].apply(lambda x : avg_ceil(x[0],x[1],x[2]),axis=1)
```

```python
# 문제
# 사용자 정의 함수 사용
# ['diameter','height','whole_weight'] 변수 사용
# 세 변수의 합이 1이 넘으면 True, 아니면 False 출력 후 answer 변수에 저장
# abalone_df에 answer 열을  추가하고 입력.
```

```python
# 내 답
def sum_data(x,y,z):
  return (x+y+z)>1
answer=abalone_df[['diameter','height','whole_weight']].apply(lambda x : sum_data(x[0],x[1],x[2]),axis=1)
abalone_df['answer']=answer
abalone_df
```

```python
# 풀이
def f(x):
  sum_=x[0]+x[1]+x[2]
  if sum_>1:
    return True
  else:
    return False

answer=abalone_df[['diameter','height','whole_weight']].apply(lambda x : f(x),axis=1)

abalone_df['answer']=answer
abalone_df
```
