```python
# dataframe 생성
# list와 dictionary로 data 생성
```

```python
# 데이터 딕셔너리 생성
friend_dic_list = [
    {'name' : 'John', 'age' : 25, 'job' : 'student'},
    {'name' : 'Nate', 'age' : 30, 'job' : 'teacher'}
]
```

```python
# dictionary는 key 순서 보장이 안됨.
df = pd.DataFrame(friend_dic_list)
df.head()
```

```python
# 원하는대로 colomn 순서 조정
df = df[['name', 'age', 'job']]
df
```

```python
# 한번에 정렬. ordered dictionary 사용
from collections import OrderedDict
friend_ordered_dict = OrderedDict(
    [
        ('name',['John','Nate']),
        ('age',[25, 30]),
        ('job',['student','teacher'])
    ]
)
```

```python
df= pd.DataFrame.from_dict(friend_ordered_dict)
df
```

```python
# list 사용해서 생성
friend_list=[
    ['John', 25, 'student'],
    ['Kate',23,None],
    ['Nate', 30, 'teacher']
]
column_name = ['name','age','job']
```

```python
df=pd.DataFrame.from_records(friend_list, columns = column_name)
df.head()
```

```python
# 한번에
friend_list = [
    ['name', ['John','Nate']],
    ['age',[25,30]],
    ['job',['student','teacher']]
]
```

```python
# 요거는 from_items가 안되네....
# 다른 방법으로는 이렇게.
df = pd.DataFrame.from_dict(dict(friend_list))
df
```

```python
# 데이터프레임 저장하는 방법. index = True, header = True는 디폴트로 설정됨.
df.to_csv('C:/Users/youth_0619/OneDrive - 중앙대학교/바탕 화면/pandasmaster/new_friends.csv' , index = True, header = True)
```
